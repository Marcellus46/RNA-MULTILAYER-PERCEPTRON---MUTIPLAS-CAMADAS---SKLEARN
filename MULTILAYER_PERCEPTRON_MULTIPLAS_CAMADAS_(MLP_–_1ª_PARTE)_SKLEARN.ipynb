## Continuação das RNA's Multilayer Perceptron - Sklearn

```python
from sklearn.neural_network import MLPClassifier
```

### Importação das Bibliotecas Iniciais do Python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Definição de Arquivos Para a RNA

```python
# O arquivo 'bezdekIris.data' não foi encontrado. Para demonstrar a funcionalidade,
# estou usando um arquivo de amostra disponível no Colab. Por favor, substitua
# o caminho abaixo pelo seu arquivo 'bezdekIris.data' quando ele estiver disponível.
iris=pd.read_csv('/content/sample_data/california_housing_train.csv')

x=iris.iloc[:,0:4].values
```

### Definindo e Buscando Modelo Sklearn

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x = scaler.fit_transform(x)

t = iris.iloc[:,4].values

from sklearn.preprocessing import LabelEncoder

label_enconder_Espécie = LabelEncoder()

t=label_enconder_Espécie.fit_transform(t)

from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer
# Reshape t to be a 2D array (150 rows, 1 column)
t = t.reshape(-1, 1)
onehotencoder_t = ColumnTransformer([('OneHot', OneHotEncoder(), [0])], remainder='passthrough')

t = onehotencoder_t.fit_transform(t)

from sklearn.model_selection import train_test_split
x_iris_treinamento, x_iris_teste, y_iris_treinamento, y_iris_teste = train_test_split(x, t, test_size=0.75, random_state=0)
```

### Treinando e Classificando A Rede MLP Skleran

```python
redeMLP = MLPClassifier(max_iter=1000, verbose=True, tol=0.0001, solver = 'adam', activation='relu',hidden_layer_sizes=(10,10,10))

redeMLP.fit(x_iris_treinamento, y_iris_treinamento)
```

### Fazendo Previsões de Teste

```python
previsoes = redeMLP.predict(x_iris_teste)

previsoes
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_iris_teste, previsoes)
```

### Print de Amostras de Classificações, Previsões e Teste

```python
print(classification_report(y_iris_teste, previsoes))
```

# Programando uma MLP de Regressão por meio da Biblioteca sklearn - 2ª Parte

```python
from sklearn.neural_network import MLPRegressor
```

## Importando Biblioteca Python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Definindo Modelo Regressão Sklearn - Gráfico

```python
x = np.linspace(-5, 5, 100)

y=np.sin(x)*np.cos(x)

plt.plot(x,y)

plt.show()
```

## Print de Amostragem de Treinamento de Regressão sklearn

```python
redeMLP = MLPRegressor(max_iter=10000, verbose=True, tol=0.000001, solver = 'adam',hidden_layer_sizes=(10,10,10))

redeMLP.fit(x.reshape(-1, 1), y)
```

## Print de Amostragem de Previsões de Regressão sklearn

```python
previsoes = redeMLP.predict(x.reshape(-1, 1))

previsoes
```

## Busca do Modelo de regressão Sklearn

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y, previsoes)

from sklearn.metrics import r2_score

r2_score(y, previsoes)
```

## Print de 2ª Amostragem de Modelo de Regressão Sklearn

```python
plt.plot(x,y)

plt.plot(x,previsoes)

plt.show()
```
